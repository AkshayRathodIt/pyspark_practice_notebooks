```python
import findspark
findspark.init('/home/akshay/spark-3.5.1-bin-hadoop3')
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("spark_tutorials_1").getOrCreate()
```

lets.read data of parquet file

```python
df = spark.read.parquet("./data/userdata1.parquet")
```

show the df

```python
df.show()
```

```python
print(help(df.show))
```

by default it shows only 20 characters per column

truncate = false will show all columns

```python
df.show(truncate=False)
```

by default it will show only 20 rows use n= Integer for different number of rows

```python
df.show(n=100)
```

```python
df.show(vertical=True)
```
